# Read images

```python
def load_images_from_folder(folder):
    images = {}
    for filename in os.listdir(folder):
        img = cv2.imread(os.path.join(folder,filename), cv2.IMREAD_GRAYSCALE)
        if img is not None:
            images[filename.split(".")[0]]=img
    return images
```

```python
#The text files have to be in the same directory as the code file
history=list(open("user_bought_history.txt", 'r')) 
items=list(open("dim_itemsï¼ˆnew).txt", 'r')) 
```

```python
#split history columns
for i in range(len(history)):
    history[i]=history[i].split(" ")
```

```python
#split item columns
for i in range(len(items)):
    items[i]=items[i].split(" ")
```

```python
#Write a function to look up for a specific item 
def get_item_index(item_id):
    for i in range(len(items)):
        if int(items[i][0])==int(item_id):
            return i
            break
        
```

```python
#Join history with items
#This code still needs to be fixed
joined_list=[]
for i in range(len(history)):
    item_index=get_item_index(history[i][1])
    joined_list[i]=history[i]+items[item_index]
```
