```python
import os
import matplotlib.pyplot as plt

from tensorflow.keras.optimizers import Adam
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Dropout, Flatten, Dense, BatchNormalization
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
from tensorflow.keras.models import Model
from tensorflow.keras.applications.resnet import ResNet152, preprocess_input
```

```python
TRAINING_DIR = 'data/training/'
train_datagen = ImageDataGenerator(preprocessing_function=preprocess_input,
                                  rotation_range=30,
                                  zoom_range=0.4,
                                  horizontal_flip=True)
```

```python
img_width = 128; img_height=128
batch_size = 64
```

```python
train_generator = train_datagen.flow_from_directory(TRAINING_DIR,
                                                   batch_size=batch_size,
                                                    class_mode='categorical',
                                                    target_size=(img_height, img_width)
                                                    )
```

```python
VALIDATION_DIR = 'data/validation/'
validation_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
validation_generator = validation_datagen.flow_from_directory(VALIDATION_DIR,
                                                   batch_size=batch_size,
                                                    class_mode='categorical',
                                                    target_size=(img_height, img_width))
```

```python
callbacks = EarlyStopping(monitor='val_loss', patience=4, verbose=1, mode='auto')
best_model_file = 'data/resnet152_drop_batch_best_weights_256.h5'
best_model = ModelCheckpoint(best_model_file, monitor='val_accuracy', verbose=1, save_best_only=True)
```

```python
# Set pretrained weight
#wp = 'data/resnet101_weights_tf.h5'
# Build base model
resnet152_base = ResNet152(include_top=False, weights='imagenet',
                          input_tensor=None, input_shape= (img_height, img_width, 3))
```

```python
print('Adding new layers...')
output = resnet152_base.get_layer(index= -1).output
output = Flatten()(output)
output = Dense(1024, activation = 'relu')(output)
output = BatchNormalization()(output)
output = Dropout(0.2)(output)
output = Dense(128, activation = 'relu')(output)
output = BatchNormalization()(output)
output = Dropout(0.2)(output)
output = Dense(15, activation='softmax')(output)
print('New layers added!')
```

```python
resnet152_model = Model(resnet152_base.input, output)
for layer in resnet152_model.layers[:20]:
    layer.trainable = False
    
resnet152_model.summary()
```

```python
resnet152_model.compile(optimizer='Adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
history = resnet152_model.fit_generator(train_generator, epochs=50, verbose=1, 
                                       validation_data = validation_generator,
                                       callbacks = [callbacks, best_model])
```

```python
target_dir = 'data'
if not os.path.exists(target_dir):
    os.mkdir(target_dir)

resnet152_model.save(target_dir + 'resnet152_model.h5')
resnet152_model.save_weights(target_dir + 'resnet152_weights.h5')

accuracy = history.history['accuracy']
val_accuracy = history.history['val_accuracy']
loss = history.history['loss']
val_loss = history.history['val_loss']
epochs = range(len(accuracy))
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(epochs, accuracy, 'r', label='Training Accuracy')
plt.plot(epochs, val_accuracy, 'b', label='Validation Accuracy')
plt.xlabel('Epoch')
plt.ylabel('Accuracy')
plt.title('Training and Validation Accuracy of ResNet152')
plt.legend(loc='lower right')
plt.show()
fig.savefig('data/Accuracy_curve_resnet152_drop_batch_256.jpg')
```

```python
fig = plt.figure(figsize=(20,10))
plt.plot(epochs, loss, 'r', label='Training Loss')
plt.plot(epochs, val_loss, 'b', label='Validation Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss of ResNet152')
plt.legend(loc='upper right')
plt.show()
fig.savefig('data/Loss_curve_resnet152_drop_batch_256.jpg')
```
